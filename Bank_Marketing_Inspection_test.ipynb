# Bank Marketing Data Analysis
This notebook presents an analysis of the `bankmarketing.csv` dataset, which includes data related to a bank's marketing campaigns. The main goal is to understand customer behavior and predict whether a client will subscribe to a term deposit.

```python

import pandas as pd

# Load the dataset
df = pd.read_csv('bankmarketing.csv')

# Display the first ten rows
df.head(10)
```

```python

# Check for missing values and data types
df.info()
df.isnull()


```

```python

# Summary statistics
print("describe:\n",df.describe())
count=df['duration'].count()
mean=df['duration'].mean()
median=df['duration'].median()
mode=df['duration'].mode()
std=df['duration'].std()
var=df['duration'].var()
min=df['duration'].min()
max=df['duration'].max()
quan1=df['duration'].quantile(0.25)
quan2=df['duration'].quantile(0.5)
quan3=df['duration'].quantile(0.75)

print("Count:",count)
print("Mean:",mean)
print("Median:",median)
print("Mode:",mode)
print("Standard Deviation:",std)
print("Variance:",var)
print("Minimum:",min)
print("Maximum:",max)
print("Quantile of(0.25)",quan1)
print("Quantile of(0.5)",quan2)
print("Quantile of(0.75)",quan3)





```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of the target variable
sns.countplot(x='y', data=df)
plt.title('Distribution of Target Variable')
plt.show()

sns.displot(df['duration'])
plt.title('Distribution of Duration')
plt.show()

sns.histplot(df['loan'])
plt.title('Distribution of loan')
plt.show()




```

```python

# Correlation matrix for numerical features

correlation=df['duration'].corr(df['campaign'])
print(correlation)
```
